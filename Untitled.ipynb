```python
from graphviz import Source
from sklearn.tree import export_graphviz
from math import sqrt
import matplotlib.pyplot as plt
from warnings import catch_warnings
from warnings import filterwarnings
from pandas import DataFrame
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.linear_model import HuberRegressor, Lars, LassoLars
from sklearn.linear_model import PassiveAggressiveRegressor, RANSACRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import ExtraTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
%matplotlib inline

def main():
    df = DataFrame({'x': [10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0, 100.0, 10.0, 20.0, 30.0, 40.0,
                          50.0, 60.0, 70.0, 80.0, 90.0, 100.0, 10.0, 20.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, 90.0, 100.0]})
    n_test = 4
    graph = MachineLearningSpotCheck().execute(df, n_test)
    return graph

class MachineLearningSpotCheck:

    def execute(self, df, n_test):
        """
        :param df: The entire dataset
        :param n_test: The test set size
        :returns: returns the sorted model scores
        """
        df = self._add_lags_y_and_drop_na(df)
        X = df.drop(['y'], axis=1).values
        y = df['y'].values
        model = DecisionTreeRegressor()
        model.fit(X, y)
        dot_data = export_graphviz(model, out_file=None)
        graph = Source(dot_data)
        graph.render("iris")        
        return graph
        

    def _add_lags_y_and_drop_na(self, df):
        """
        :param df: The entire dataset
        :returns: returns the df with 6 lags, y, and drops NAs.
        """
        for i in range(1, 7):
            df.loc[:, 'lag_' + str(i)] = df['x'].shift(i)
        df.loc[:, 'y'] = df['x'].shift(-1)
        df = df.dropna()
        return df

main()
```
